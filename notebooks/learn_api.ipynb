# Sessao 1 -  HTTP na Prática

```python
import requests
```

```python
response = requests.get("https://viacep.com.br/ws/01310100/json/")
```

```python
print(response.status_code)
print(type(response.json()))
print(response.json())
```

- O que é status_code 200? E 404? E 422?

    - São codigos para dizer se a requisição foi bem sucedida ou nao. Cada codigo tem um significado. 200 é sucesso.

- Qual a diferença entre GET e POST?
    - Get voce pega dados e post voce envia dados.

- Por que a resposta vem como JSON e não como DataFrame?

    - Porque por essa estrutura de dados que a api se comunica.

# Sessao 2 - Pydantic Isolado

```python
from pydantic import BaseModel

class Cliente(BaseModel):
    nome: str
    tenure: int
    monthly_charges: float


c1 = Cliente(nome="João", tenure=12, monthly_charges=29.99)
print(c1)
print(c1.model_dump())
```

```python
# Agora tentando com o formato errado
c2 = Cliente(nome="Maria", tenure="doze", monthly_charges="vinte e nove")
print(c2)
```

```python
c3 = Cliente(nome="Maria",  monthly_charges=29.99)
print(c3)
```
